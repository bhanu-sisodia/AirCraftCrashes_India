```python
df = pd.read_csv('data/ATfatalities.csv')
```

```python
#Plot the timeline with dot size and color by fatalities
sns.set(style="whitegrid")
plt.figure(figsize=(10, 2))
sns.scatterplot(
    x='Year',
    y=[1] * len(df),  # All dots on a flat horizontal line
    size='Fatalities',
    sizes=(20, 400),  # The data has a range of 18 to 349, this will allow a reasonable representation
    color = 'red',
    data=df,
    legend=False,
)

#Annotate major incidents
for i, row in df.iterrows():
    if row['Fatalities'] >= 100:
        plt.text(row['Year'], 1.04, f"{row['Year']} ({row['Fatalities']})", ha='center', fontsize=9)

#Final formatting
plt.yticks([])
plt.xlabel('Year')
plt.title('Major Airliner Accidents in India â€” Dot Size = Fatalities', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
